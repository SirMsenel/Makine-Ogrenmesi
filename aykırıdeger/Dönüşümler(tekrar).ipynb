Standartlaştırma

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])
df = pd.DataFrame(
    {
        "V1" :V1,
        "V2" : V2,
        "V3" : V3
    }
)
df = df.astype(float)
df
```

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)
##ortlama = 0 ,standart sapma =1 olacak şekilde standartlaştırma
```

 Normalizasyon

```python
preprocessing.normalize(df)
```

Min-Max Dönüşümü

```python
scaler = preprocessing.MinMaxScaler(feature_range=(10,20))
```

```python
scaler.fit_transform(df)
```

Binarize Dönüşüm(0-1 dönüşümü(belirli bir eşikteki değerleri))

```python
binarizer = preprocessing.Binarizer(threshold=5).fit(df)
```

```python
binarizer.transform(df)
```
